```python
import pandas as pd
from kafka import KafkaConsumer
from time import sleep
from json import dumps,loads
import json
from s3fs import S3FileSystem
```

```python
# pip install s3fs
```

```python
consumer = KafkaConsumer('demo-test',
                         bootstrap_servers=['18.191.170.67:9092'],
                         value_deserializer= lambda x: loads(x.decode('utf-8')))
                        
```

```python
s3= S3FileSystem()
```

```python
for count, i in enumerate(consumer):
    print(i.value)
    with s3.open("s3://kafka-stock-market-flow/stock_market_{}.json".format(count),'w') as file:
        json.dump(i.value,file)
    
```
